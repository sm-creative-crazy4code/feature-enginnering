````python
import pandas as pd
import numpy as np
````

````python
data = pd.read_csv("train.csv",usecols=['X1','X2','X3','X4','X5','X6'])
````

````python
data.head()
````

``` here we will try to find out how many unique categories are present for each feature ```

``` here the .unique() function only gives those attributes that have unique values  ```

````python
for col in data.columns:
    print(col,":",len(data[col].unique())," labels")
````

###### performing one hot encoding

````python
pd.get_dummies(data,drop_first=True).shape
````

``` here basically we can observe that for 6 initial features we are getting 117 columns ie 112 extra so
as the no of features increase the no of dimensions will increase which will lead to curse of dimensionality```

```hence indstead of seclecting the all feature we will one hot encoding on the most frequent occuring feature that is in each category for
some most frequently occuring varible we will create one binary varible```

````python
data.X2.value_counts().sort_values(ascending=False ).head(20)
````

```  top 10 features that is from as to  e will be 1 and rest will be 0```

````python
top_10=[x for x in data.X2.value_counts().sort_values(ascending=False ).head(10).index ]
top_10
````

```making top 10 binary varibles ```

````python
for lable in top_10:
    data[lable]=np.where(data['X2']==lable,1,0)
                         
data[['X2']+top_10].head(40)
````

``` getting a whole new set of dummy variables for each categorical varible ```

````python
def one_hot_top_x(df, variable, top_x_lables):
    for lable in top_x_lables:
        df[variable+' '+lable]=np.where(data['X2']==lable,1,0)
        
        
````

````python
data = pd.read_csv("train.csv",usecols=['X1','X2','X3','X4','X5','X6'])
````

````python
one_hot_top_x(data,'X2',top_10)
````
